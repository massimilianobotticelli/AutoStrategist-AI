```python
df_original = spark.table('workspace.car_sales.vehicles')
```

```python
df_cleaned = df_original.filter(df_original.manufacturer.isin('ford', 'toyota', 'chevrolet', 'honda')).sample(0.01)
```

```python
df_cleaned = df_cleaned.drop("url", "region", "region_url", "VIN", "image_url", "county", "size", "title_status", "state", "lat", "long", "posting_date")
```

```python
df_cleaned = df_cleaned.filter((df_cleaned.price != 0) & df_cleaned.year.isNotNull() & df_cleaned.model.isNotNull() & df_cleaned.odometer.isNotNull() & df_cleaned.price.isNotNull())
```
